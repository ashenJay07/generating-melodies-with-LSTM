```python
from music21 import note, stream

# Create a simple music score
score = stream.Score()
note = note.Note("C4")
score.append(note)

# Check if the score is well-formed
if not score.isWellFormedNotation():
    print("The score is not well-formed.")
```

```python
from music21 import stream, note, meter

# Create a simple music score with measures and staves
score = stream.Score()
part = stream.Part()
measure = stream.Measure()
measure.append(meter.TimeSignature('4/4'))
measure.append(note.Note("C4"))
part.append(measure)
score.insert(0, part)
```

```python
score.show("musicxml")  # open with musescore application
```
